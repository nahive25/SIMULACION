```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Generar 100 números aleatorios uniformemente distribuidos entre 0 y 1
random_numbers = np.random.rand(100)

# Realizar la prueba de Kolmogorov-Smirnov
D, p_value = stats.kstest(random_numbers, 'uniform')

# Mostrar los resultados
print(f"Estadístico D: {D}")
print(f"Valor p: {p_value}")

# Interpretación del resultado
if p_value > 0.05:
    print("No se rechaza la hipótesis nula: los números siguen una distribución uniforme")
else:
    print("Se rechaza la hipótesis nula: los números no siguen una distribución uniforme")

# Graficar el histograma en color lila
plt.hist(random_numbers, bins=10, edgecolor='black', color="#E6A4B4")
plt.title("Histograma de los números aleatorios generados")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```
