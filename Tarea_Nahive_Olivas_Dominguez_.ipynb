```python
import matplotlib.pyplot as plt

def metodo_congruencial_lineal(x0, a, c, m, n):
    numeros = []
    xi = x0
    for _ in range(n):
        xi = (a * xi + c) % m
        numeros.append(100 + (xi / m) * (450 - 100))  # Normalizamos a [100,450]
    return numeros

# Parámetros del método congruencial lineal
x0 = 7  # Semilla
a = 1664525  # Multiplicador
c = 1013904223  # Incremento
m = 2**32  # Módulo
n = 100  # Cantidad de números aleatorios

numeros_generados = metodo_congruencial_lineal(x0, a, c, m, n)

# Mostrar los primeros 10 números generados
print("Primeros 10 números generados:", numeros_generados[:10])

# Graficar la distribución de los números generados
plt.hist(numeros_generados, bins=10, edgecolor='black', color='pink')
plt.xlabel('Valor generado')
plt.ylabel('Frecuencia')
plt.title('Histograma de los números generados')
plt.show()
```
